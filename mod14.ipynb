***RNN***

```python
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences

max_features = 10000
maxlen = 200
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# RNN
model_rnn = Sequential()
model_rnn.add(Embedding(input_dim=max_features, output_dim=128, input_length=maxlen))
model_rnn.add(SimpleRNN(64))
model_rnn.add(Dense(1, activation='sigmoid'))

model_rnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_rnn = model_rnn.fit(
    x_train, y_train, epochs=5, batch_size=32, validation_split=0.2, verbose=1
)

loss_rnn, accuracy_rnn = model_rnn.evaluate(x_test, y_test)
print(f"Test Loss (RNN): {loss_rnn:.4f}")
print(f"Test Accuracy (RNN): {accuracy_rnn:.4f}")
```

***LSTM 1 LAYER***

```python
from keras.layers import LSTM

# LSTM
model_lstm = Sequential()
model_lstm.add(Embedding(input_dim=max_features, output_dim=128, input_length=maxlen))
model_lstm.add(LSTM(64))
model_lstm.add(Dense(1, activation='sigmoid'))

model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_lstm = model_lstm.fit(
    x_train, y_train, epochs=5, batch_size=32, validation_split=0.2, verbose=1
)

loss_lstm, accuracy_lstm = model_lstm.evaluate(x_test, y_test)
print(f"Test Loss (LSTM): {loss_lstm:.4f}")
print(f"Test Accuracy (LSTM): {accuracy_lstm:.4f}")
```

***Двостороння***

```python
from keras.layers import Bidirectional

# Bi-LSTM
model_bi_lstm = Sequential()
model_bi_lstm.add(Embedding(input_dim=max_features, output_dim=128, input_length=maxlen))
model_bi_lstm.add(Bidirectional(LSTM(64)))
model_bi_lstm.add(Dense(1, activation='sigmoid'))

model_bi_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_bi_lstm = model_bi_lstm.fit(
    x_train, y_train, epochs=5, batch_size=32, validation_split=0.2, verbose=1
)

loss_bi_lstm, accuracy_bi_lstm = model_bi_lstm.evaluate(x_test, y_test)
print(f"Test Loss (Bi-LSTM): {loss_bi_lstm:.4f}")
print(f"Test Accuracy (Bi-LSTM): {accuracy_bi_lstm:.4f}")
```

***Глибока LSTM***

```python
# Deep LSTM
model_deep_lstm = Sequential()
model_deep_lstm.add(Embedding(input_dim=max_features, output_dim=128, input_length=maxlen))
model_deep_lstm.add(LSTM(64, return_sequences=True))
model_deep_lstm.add(LSTM(64))
model_deep_lstm.add(Dense(1, activation='sigmoid'))

model_deep_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_deep_lstm = model_deep_lstm.fit(
    x_train, y_train, epochs=5, batch_size=32, validation_split=0.2, verbose=1
)

loss_deep_lstm, accuracy_deep_lstm = model_deep_lstm.evaluate(x_test, y_test)
print(f"Test Loss (Deep LSTM): {loss_deep_lstm:.4f}")
print(f"Test Accuracy (Deep LSTM): {accuracy_deep_lstm:.4f}")
```

***Графіки***

```python
import matplotlib.pyplot as plt

# Accuracy
plt.plot(history_rnn.history['val_accuracy'], label='RNN')
plt.plot(history_lstm.history['val_accuracy'], label='LSTM')
plt.plot(history_bi_lstm.history['val_accuracy'], label='Bi-LSTM')
plt.plot(history_deep_lstm.history['val_accuracy'], label='Deep LSTM')

plt.title('Validation Accuracy Comparison')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss
plt.plot(history_rnn.history['val_loss'], label='RNN')
plt.plot(history_lstm.history['val_loss'], label='LSTM')
plt.plot(history_bi_lstm.history['val_loss'], label='Bi-LSTM')
plt.plot(history_deep_lstm.history['val_loss'], label='Deep LSTM')

plt.title('Validation Loss Comparison')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

***Висновок***

**Проста RNN демонструє найгірші результати через обмежену здатність працювати з довгими послідовностями, тому вона не підходить для таких задач. LSTM показує хорошу продуктивність із точністю близько 85% і стабільним зменшенням втрат, що робить її ефективним рішенням для обробки тексту. Двостороння LSTM (Bi-LSTM) досягає найвищої точності (~87%) завдяки врахуванню контексту з обох сторін тексту, що робить її найефективнішою моделлю для цього завдання. Глибока LSTM, хоча й демонструє схожі результати з LSTM, має ризик перенавчання через складність моделі, що проявляється у зростанні втрат на пізніх епохах. Загалом, Bi-LSTM є оптимальним вибором для задач аналізу тексту, тоді як LSTM підходить для випадків із обмеженими ресурсами.**
